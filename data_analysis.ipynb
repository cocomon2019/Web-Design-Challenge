```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the csv
df = pd.read_csv('Resources/cities.csv')
df.head()
```

```python
#plot the graph
plt.scatter(df['Lat'], df['Max Temp'],edgecolors='black',alpha=0.75)
plt.grid()

plt.ylabel('Max Temperature (F)')
plt.xlabel('Latitude')
plt.title('City Latitude vs Max Temperature (11/30/2019)')

plt.tight_layout()

#save fig
plt.savefig("City Latitude vs Max Temperature")
plt.show()
```

```python
#plot the grpah
plt.scatter(df['Lat'], df['Humidity'],edgecolors='black',alpha=0.75)
plt.grid()

plt.ylabel('Humidity (%)')
plt.xlabel('Latitude')
plt.title('City Latitude vs Humidity (11/30/2019)')

plt.tight_layout()

#save fig
plt.savefig("City Latitude vs Humidity")
plt.show()
```

```python
#plot the graph
plt.scatter(df['Lat'],df['Cloudiness'],edgecolors='black',alpha=0.75)
plt.grid()

plt.ylabel('Cloudiness (%)')
plt.xlabel('Latitude')
plt.title('City Latitude vs Cloudiness (11/30/2019)')

plt.tight_layout()

#save fig
plt.savefig("City Latitude vs Cloudiness")
plt.show()
```

```python
#plot the graph 
plt.scatter(df['Lat'],df['Wind Speed'],edgecolors='black',alpha=0.75)
plt.grid()

plt.ylabel('Wind Speed (mph)')
plt.xlabel('Latitude')
plt.title('City Latitude vs Wind Speed (11/30/2019)')

plt.tight_layout()

#save fig
plt.savefig("City Latitude vs Wind Speed")
plt.show()
```
